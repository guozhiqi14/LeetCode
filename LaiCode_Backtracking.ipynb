### Print out Subset   
Non-duplicate case

```python
def subset(num):
    res,cur = [],[]
    backtrack(num,0,res,cur)
    return res

def backtrack(num,index,res,cur):
    if index == len(num):
        res.append(cur[:])
        return
    cur.append(num[index])
    backtrack(num,index+1,res,cur)
    cur.pop()
    backtrack(num,index+1,res,cur)
    
subset(['a','b'])
```

### Print all valid combinations of factors that form an integer.

```python
import math
class Solution:
    def getFactors(self, n):
        """
        :type n: int
        :rtype: List[List[int]]
        """
        solution,res = [],[]
        def getFactorsHelper(n,index,solution,res):
            if n == 1:
                res.append(solution[:])
                return 
            for i in range(index,n+1):
                if n%i==0:
                    solution.append(i)
                    getFactorsHelper(n//i,i,solution,res)
                    solution.pop()
                    
        getFactorsHelper(n,2,solution,res)
        res.pop()
        return res
       
```

```python
s = Solution()
s.getFactors(12)
```

### Generate Valid Parentheses

Given n pairs of parentheses, write a function to generate all combinations of well-formed parentheses.

For example, given n = 3, a solution set is:

```python
def generateParenthesis(n):
    cur = []
    res = []
    helper(0,0,n,cur,res)
    return res
def helper(left,right,n,cur,res):
    if left == n and right == n:
        res.append(''.join(cur))
        return 
    if left < n:
        cur.append('(')
        helper(left+1,right,n,cur,res)
        cur.pop()
    if right < left:
        cur.append(')')
        helper(left,right+1,n,cur,res)
        cur.pop()
```

### Follow up

what if we have 3 kinds of parentheses({},<>,())   
Idea: we need to keep track of last unmatched left parentheses.

```python
def validParentheses(l,m,n):
    remain = [l,l,m,m,n,n]
    parentheses = ['(',')','<','>','{','}']
    targetlen = 2*(l+m+n)
    cur,stack,res = [],[],[]
    bt(cur,stack,remain,targetlen,res,parentheses)
    return res
def bt(cur,stack,remain,targetlen,res,parentheses):
    if len(cur) == targetlen:
        res.append(''.join(cur[:]))
        return
    for i in range(len(remain)):
        if i%2==0:
            if remain[i]>0:
                cur.append(parentheses[i])
                stack.append(parentheses[i])
                remain[i] -= 1
                bt(cur,stack,remain,targetlen,res,parentheses)
                cur.pop()
                stack.pop()
                remain[i] += 1
        else:
            if stack and stack[-1] == parentheses[i-1]:
                cur.append(parentheses[i])
                stack.pop()
                remain[i]-=1 
                bt(cur,stack,remain,targetlen,res,parentheses)
                cur.pop()
                stack.append(parentheses[i-1])
                remain[i]+=1
validParentheses(2,2,2)
```
